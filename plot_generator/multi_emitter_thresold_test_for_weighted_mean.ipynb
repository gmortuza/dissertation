```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

multi_emitter_file_name = "data/multi_emitter_threshold_test_for_weighted_mean.xlsx"
excel_file = pd.ExcelFile(multi_emitter_file_name)
data = excel_file.parse(excel_file.sheet_names)
```

```python
# Remove Nan
# sort
for sheet_name, df in data.items():
    # Remove nan and sort
    data[sheet_name] = df.dropna().sort_values(by=['Threshold'], ignore_index=True)
```

```python
matplotlib.rc('figure', figsize=(10, 5))
plt.rcParams['font.size'] = 14
plt.rcParams["font.family"] = "Times New Roman"
for sheet_name in data.keys():
    label_name = r" = " + sheet_name.split("_")[-1]
    plt.plot(data[sheet_name]['Threshold'], data[sheet_name]['efficiency'], label=label_name)
    # Find the maximum value
    max_coordinate = data[sheet_name]['efficiency'].argmax()

    plt.scatter(data[sheet_name]['Threshold'][max_coordinate], data[sheet_name]['efficiency'][max_coordinate], marker="^")
plt.xlabel("Threshold")
plt.ylabel("Efficiency (%)")
plt.legend()
plt.savefig("multi_emitter_threshold_test_weighted_mean.png", dpi=300, bbox_inches="tight")
plt.show()
```
